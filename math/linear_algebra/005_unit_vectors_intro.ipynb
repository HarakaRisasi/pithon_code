$\hat{i} - Unit \ vector$

Единичный вектор, или орт, — вектор нормированного пространства, длина которого равна единице. Единичные вектора используются, в частности, для задания направлений в пространстве. Множество единичных векторов образует единичную сферу. Единичный вектор часто обозначается строчной буквой с крышкой.

---
Example:

$
\overrightarrow{V} =
\begin{align}
\begin{bmatrix}
2\\
3
\end{bmatrix}
\end{align}
$

Unit vector:

$
\hat{i} =
\begin{align}
\begin{bmatrix}
1\\
0
\end{bmatrix}
\end{align}
$
$
\hat{j} =
\begin{align}
\begin{bmatrix}
0\\
1
\end{bmatrix}
\end{align}
=>
\overrightarrow{V} = 2\hat{i} + 3\hat{j}
$




---

$
\overrightarrow{b} = -1\hat{i} + 4\hat{j} = 
\begin{align}
\begin{bmatrix}
-1\\
4
\end{bmatrix}
\end{align}
$

$
\overrightarrow{V} + \overrightarrow{b} = 1\hat{i} + 7\hat{j} = 
\begin{align}
\begin{bmatrix}
1\\
7
\end{bmatrix}
\end{align}
$
