### О чем этот курс и как он устроен

Курс "Введение в данные" — первый в специализации "Анализ данных", разработанной Новосибирским государственный университетом совместно с компанией 2GIS.

В этом курсе вы **изучите (или вспомните)** `основы теории вероятности` и `математической статистики`, `графического анализа данных`, а также узнаете основные `виды выборок и ограничения, с ними связанные`. Мы покажем, `как рассчитывать описательные статистики` и `строить графики в средах анализа данных (SPSS и R)`, а вы сделаете первые расчеты.

**Первый модуль курса посвящён основам теории вероятностей**. Если вы не знакомы с теорией вероятностей или успели её подзабыть, мы рекомендуем отнестись к этому модулю серьезно: понимание теории вероятностей необходимо для успешного прохождения как этого курса, так и всей специализации. Дальше мы рассмотрим основы математической статистики, узнаем об основных характеристиках распределений, особенностях разных статистических показателей и способах их расчёта. Мы также научимся строить графики, формировать выборки, работать с пропущенными данными и кое-что посчитаем.

- Какими бывают данные?
- Как кодировать и хранить данные?
- Как рассчитать вероятность наступления события?
- Как тип данных ограничивает нас в применении статистических процедур?
- Как, имея данные, определить границы типичных и нетипичных значений?
- Какие существуют графические инструменты анализа и как выбрать график, наиболее подходящий для нашей задачи?
- Как сформировать хорошую выборку и как рассчитать ее объём?
- Что делать с пропущенными и неопределенными данными?
- Эти и другие вопросы мы обсудим в данном курсе.

**В конце второго, третьего и четвертого модулей** есть практические видео, которые пошагово демонстрируют решение задач в SPSS и в R.

Каждый модуль завершается **оцениваемым тестом**, который позволяет проверить и применить полученные знания.

В разделе "**Данные**" вы найдете ссылки на массивы данных, с которыми мы работаем и на которые ссылаемся в лекциях, и их краткое описание.

В разделе "**Дополнительные материалы по статистическим пакетам**" находятся материалы по средам, в которых мы показываем примеры анализа данных (SPSS и R). Там вы также найдёте ссылки, по которым можно скачать и установить софт на свой компьютер и попрактиковаться. Кроме того, туда же мы будем выкладывать материалы и учебники, на которые ссылаемся.

> И наконец к каждой лекции есть соответствующий форум, где можно обсудить материал и задать вопросы, если что-то осталось непонятным.
