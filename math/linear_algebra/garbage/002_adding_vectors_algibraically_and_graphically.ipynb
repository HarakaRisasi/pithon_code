```python
# Using SymPy and Numpy

# Import
# --------------------------------------------------------
import sympy as sp
import numpy as np

from sympy.interactive import printing
printing.init_printing(use_latex=True)
from sympy import Eq, solve_linear_system, Matrix, symbols, bell
from numpy import linalg
# --------------------------------------------------------

# Sympy
print('Using Sympy: ')

eq1 = sp.Function('eq1')
eq2 = sp.Function('eq2')

x,y  = sp.symbols('x y')

eq1 = Eq(2*x-y, -4)
eq2 = Eq(3*x-y, -2)

display(eq1)
display(eq2)

row1=[2,-1,-4]
row2=[3,-1,-2]

system = Matrix((row1, row2))

display(system)

display(solve_linear_system(system, x, y))

print("================================")

# Numpy
print("Using Numpy:")

nrow1 = [2, -1]
nrow2 = [3, -1]

nmat = np.array([nrow1, nrow2])
cons = np.array([-4, -2])

answer = linalg.solve(nmat, cons)
xval = int(answer[0])
yval = int(answer[1])

print(f" x = {xval}\n y = {yval} ")
```
