### Фильтр данных компания\регион.

```python
def filter_region(index_region):
    """Функция фильтрует СЦ по регионам"""
    # Упорядочивание по рерионам
    list_region = catalog['Наименование региона филиала'].unique().tolist()
    # Новый фрейм с регионами
    region = pd.DataFrame({'Регион':list_region})
    # region


    # Получение списка(Кодов) филиалов конкретного региона >>> для получения другого списка, нужно изменить индекс региона.
    df_identificator = catalog['Идентификатор филиала'].loc[catalog['Наименование региона филиала'] == region['Регион'][index_region]]
    list_identificator = df_identificator.to_list()
    # list_identificator


    # Фильтр по регионам.
    frames = []
    for i in range(len(list_identificator)):
        frames += [df.loc[df['Идентификатор_филиалы_документа'] == list_identificator[i]]]
    df_region = pd.concat(frames)

    return df_region


def dynamics_company():
    """Функция вычисляет динамику среднего время нахождения товара в сервисном центре по всей компании"""
    week = []
    
    for i in range(1, df['Неделя_x'].unique().max() + 1):
        week.append(round(df['Срок гарантийного обслуживания'].loc[df['Неделя_x'] == i].mean(), 2))
        
    df_week_company = pd.DataFrame({'Неделя': df['Неделя_x'].unique(), 'Среднее значение': week})
    
    return df_week_company


def dynamics_region(index_region):
    """Функция вычисляет динамику среднего время нахождения товара в сервисном центре по региону"""
    
    # Расчет среднего кол-ва дней по неделям
    df_region = filter_region(index_region) # Наследует др.функцию.
    df_week = pd.DataFrame({ 'Неделя': df_region['Неделя_x'], 'Среднее значение': df_region['Срок гарантийного обслуживания']})
    df_week = df_week.groupby('Неделя').mean().reset_index()

    return df_week
```

```python
df_week_company = dynamics_company()
```

а) В феврале замечается снижение количества заявок обрабатываемых до 30 дней.

Здесь важно отметить, что кол-во товаров(заявок) находящихся в СЦ в период до 15 дней значительно снижается в отличии от других месяцев, а в период от 16 до 31 дня падает до 0.

***Заключение***: Основываясь на полученных статистических данных, можно вывести три нормативных срока по всей компании:
- 1) от 0 до 15 дней - для простого ремонта.
- 2) от 16 до 30 дней - для ремонта средней тяжести при наличии зап.частей.
- 3) от 30 до 45 дней - для случаев где требуется серьезный и длительный ремонт, а так же заказ зап.частей.

### Графическая динамика отклонения от нормы по каждому региону.

```python
# Регион
# 0	Верхняя Волга
# 1	Центральный
# 2	Черноземье
# 3	Приволжский
# 4	Средняя Волга
# 5	Урал
# 6	Западная Сибирь
# 7	Юг
# 8	Дальний Восток
# 9	Восточная Сибирь

# Для получения отчета введите индекс из списка регионов.
region_index = 0 # <<<

df_week_region = dynamics_region(region_index)
```

### Графики

```python
# Получение данных по среднему времени нахождения товара в сервисном центре (вся компания, регион).

def deviation(x):
    """Функция помогает найти разницу между нормативами (вся компания\регион)"""
    x.mean()
    return norm - x

region_deviation = round(deviation(dynamics_region(region_index)['Среднее значение']).mean(), 2) # среднее отклонение(дней) по региону от нормы всей компании
print(f'Cреднее отклонение по региону от нормы всей компании (*дней) = {region_deviation}')
```

```python
# Фрейм с данными о среднем времени нахождения товара в сервисном центре в разрезе (вся компания, регион)
df_norma_region = pd.DataFrame({"Норма ВсяКомпания\Регион": [norm, region_deviation+norm, region_deviation]}, 
                               index=['ВсяКомпания',f'{list_region[region_index]}', 'Отклонение(*дней)'])

# Отрисовка горизонтального графика.
x = ["Вся компания", f"{list_region[region_index]}"]
y = [df_norma_region['Норма ВсяКомпания\Регион'][0], df_norma_region['Норма ВсяКомпания\Регион'][1]]
ticks = np.arange(int(df['Срок гарантийного обслуживания'].max()))

fig, ax = plt.subplots(figsize=(16, 4), dpi=100)
plt.barh(x, y)

# Значения для каждого из Бинов.
for index, value in enumerate(y):
    plt.text(value, index, str(value))
    
plt.title(f'Отклонение от общей нормы региону (за Год) {list_region[region_index]}')
plt.xticks(ticks)
plt.xlabel('Дней в СЦ (*Средняя)')
plt.show();

df_norma_region
```

### Вычисление стандартного отклонения от нормы по всем регионам.

```python
# Вычисление стандартного отклонения от нормы по всем регионам.
d = {}

for i in range(len(list_region)):
    
    # Получение отклонения от среднего значения понедельно за год по региону.
    def standard_deviation_region():
        """Этапы вычисления стандартного отклонения с преминением к Среднему сроку нахождения товара в разрезе регионов"""
        def step_one(x):
            j = x - norm # Для каждого числа в наборе данных вычитаю его со средним по всей компании.
            return j ** 2 # Возвожу в квадрат разницу.

        step_two = dynamics_region(i)['Среднее значение'].apply(step_one).sum() # Сложение итогов шага 1.

        step_three = step_two / (len(dynamics_region(i)-1)) # Разделение суммирование из шага 2 на количество значений в наборе данных минус один.

        return np.sqrt(step_three) # Квадратный корень - результат шага 3.
    
    d[list_region[i]] = round(standard_deviation_region(), 2) # Составление словаря, с полученными данными.
```

```python
# Отображение данных в виде ДатаФрейма.
full_std_dev = pd.DataFrame({"Станд.откл от общей нормы": d})
full_std_dev.reset_index(inplace=True)
full_std_dev = full_std_dev.rename(columns = {'index':'Регион'})
full_std_dev
```
