$\hat{i} - Unit \ vector$

Единичный вектор, или орт, — вектор нормированного пространства, длина которого равна единице. Единичные вектора используются, в частности, для задания направлений в пространстве. Множество единичных векторов образует единичную сферу. Единичный вектор часто обозначается строчной буквой с крышкой.

---
Example:

$
\overrightarrow{V} =
\begin{align}
\begin{bmatrix}
2\\
3
\end{bmatrix}
\end{align}
$

Unit vector:

$
\hat{i} =
\begin{align}
\begin{bmatrix}
1\\
0
\end{bmatrix}
\end{align}
$
$
\hat{j} =
\begin{align}
\begin{bmatrix}
0\\
1
\end{bmatrix}
\end{align}
=>
\overrightarrow{V} = 2\hat{i} + 3\hat{j}
$

Example: **Select the unit vector in the direction of**  $\vec{v}=(-2, 3)$

1) 

A unit vector $\hat i$ has a magnitude (or length) of 1

Dividing $\vec{v}$ by its magnitude will find a vector in the same direction as $\vec{v}$ but with a magnitude of 1:

Unit vector in the direction of $\vec{v} =  \frac{\vec{v}}{||\vec{v}||}$

2)

Finding the unit vector:

$
\frac{\vec{v}}{||\vec{v}||} = 
\frac{(-2, 3)}{\sqrt{(-2)^2 + 3^2}} = 
\frac{1}{\sqrt{13}} * (-2, 3) = 
(-\frac{2}{\sqrt{13}}, \frac{3}{\sqrt{13}})
$

Great, we found the unit vector!

3)
Verifying that the magnitude is 1:

$
\sqrt{(-\frac{2}{\sqrt{13}})^2 + (\frac{3}{\sqrt{13}})^2} = 
\sqrt{\frac{4}{13} + \frac{9}{13}} = 
\sqrt{\frac{13}{13}} = 
1
$
