## Standard deviation

1) Find $\mu$(mean):\
$$\mu = \frac{\sum_{x_i}}{n}$$

2) Find standard diviation ($\sigma$):\
$$\sqrt{\frac{\sum(x_i - \mu)^2}{n}} = \sigma$$



## Example:
{3, 4, 6, 7}

$\mu$ = 20 / 4 = 5

$\sigma = \sqrt\frac{(3-5)^2 \ + \ (4-5)^2 \ + \ (6-5)^2 \ + \ (7-5)^2}{4} = \sqrt{\frac{10}{4}} = \sqrt{2.5}$\
$\Rightarrow$\
$\sigma = 1.58$


## Objective:
> Calculate and print the standard deviation.

**Sample Input**:\
        5\
        10 40 30 50 20
        
**Sample Output**:\
        14.1

```python
from statistics import mean
from math import sqrt

n = int(input())
x = list(map(int, input().split()))
sD = 0

# find the mean
mean = mean(x)

# find the standard deviation
for i in range(n):
#     squared distance
    sD += (x[i] - mean)**2
#     standard diviation
    stD = sqrt(sD / n)
    
print(round(stD, 1))
```

## or Numpy

```python
import numpy as np

print(round(np.std(x), 1))
```
