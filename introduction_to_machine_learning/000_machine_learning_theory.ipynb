>**Что такое random_state?**\
В некоторых заданиях мы просим вас зафиксировать некоторые параметры для воспроизводимости результата. random_state — один из таких параметров. Дело в том что в алгоритмах может использоваться генератор случайных чисел, поэтому, запустив одинаковый код дважды, может получиться разныерезультаты. \
Чтобы избежать этого, мы просим зафиксировать random_state, благодаря которому мы будем уверены, что независимо от вашей \
операционной системы и версий библиотек результат останется воспроизводимым.

**Весь курс машинного обучения - это конкретизация**:\
    a. Как задаются объекты и какими могут быть ответы;\
    b. Как строить функцию $a$;\
    c. В каком смысле $a$ должен приближать $y$.

a. Как задаются объекты - самый распространенный способ задания объектов - это *признаковое описание*.

**Признаки** чисто формально - это функции, которые объектам ставят в соответствие какие-то значения, как правило, числовые.

$f_j: X \to D_j, j = 1, ..., n$ - Признаки объектов(features).

Тоесть (признаки), это какие-то способы измерения над объектами, и в зависимости от того, что это за измерения, признаки делятся на типы:\
    - $D_j = \{0,1\}$ - Бинарный признак $f_j$;\
    - $|D_j| < \infty$ - Номинальный признак $f_j$;\
    - $|D_j| < \infty, D_j - упорядочено$ - порядковый признак $f_j$;\
    - $D_j = \mathbb{R}$ - количественный признак $f_j$.
    
<div class="alert alert-block alert-info"; white-space: pre>   
    
**Бинарный признаки** - это какой-то ответ "Да" или "Нет" про интересующий нас объект.<br />
**Номинальный признаки** - тоже принимают конечное множество значений, но большее чем у *бинарных признаков*.<br />
**Порядковые признаки** - это когда на множестве значений признаков задано некое отношение порядка.<br />
**Количественные признаки** - которые являются числовыми измерениями над нашими объектами.

    *"Цвет глаз" - относится к номинальному типу признаков.
    *"Воинское звание" - - относится к порядковому типу признаков.
</div>

**Матрица "объекты - признаки"**  -
$
F  = ||f_j(x_i)||_{j*n} = 
\begin{bmatrix}
f_1(x_1) & \cdots & f_n(x_n) \\
\vdots & \cdots & \vdots \\
f_1(x_l) & \cdots & f_n(x_l)
\end{bmatrix}
$\
В этой матрице:\
\*Строки - это объекты (а так же каждой строке соответствует правильный ответ).\
\*Столбцы - это признаки.
