$
\overrightarrow{a} = 
\begin{align}
    \begin{bmatrix}
    -1 \\
    2
    \end{bmatrix}
\end{align}
\quad
$
$
\overrightarrow{b} = 
\begin{align}
    \begin{bmatrix}
    3 \\
    1
    \end{bmatrix}
\end{align}
$ 

$
\overrightarrow{a} + \overrightarrow{b} = 
\begin{align}
    \begin{bmatrix}
    -1 + 3 \\
    2 + 1
    \end{bmatrix} =
    \begin{bmatrix}
    2 \\
    3
    \end{bmatrix}
\end{align} 
$
